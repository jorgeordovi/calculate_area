```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=sp.symbols('x')

f = -x**2 + 3*x + 2
```

```python
f_lambdified=sp.lambdify(x,f,'numpy')
x_vals=np.linspace(-2,4,400)

y_vals=f_lambdified(x_vals)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x)=-x**2 + 3x + 2$', color='blue')

plt.title('Funcion')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
F_indefinida = sp.integrate(f,x)

sp.pretty_print(F_indefinida)
```

definir limites de integracion 

```python
a,b=0,3
area_definida=sp.integrate(f,(x,a,b))

print(f"Area bajo la curva entre x={a} y x={b}: {area_definida}")
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x)=-x**2 + 3x + 2$', color='blue')
x_fill=np.linspace(a,b,100)
y_fill=f_lambdified(x_fill)
plt.fill_between(x_fill,y_fill, color='orange', alpha=0.2,label=f'Area={area_definida}')
plt.title('Funcion')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
